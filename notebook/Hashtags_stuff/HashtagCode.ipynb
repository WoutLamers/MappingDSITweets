```python
df = df[df['entities'].isnull() == False]
```

```python
hashtaglist = []
for i, row in df.iterrows():
    
    hashtags = row['entities']['hashtags']
    for tag in hashtags:
        hashtaglist.append((row['index'], tag['text']))
```

```python
hashtagdf = pd.DataFrame(hashtaglist, columns=['index', 'hashtag']).set_index('index')
```

```python
hashtagdf.head()
```

```python
hashtaglist = hashtagdf['hashtag'].values.tolist()
len(hashtaglist)
```

```python
from collections import Counter
count_hashtags = Counter(hashtaglist)
```

```python
count_hashtags.most_common()
```
