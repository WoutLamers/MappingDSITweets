```python
import os, json
import pandas as pd

path_to_json = 'C:\\Users\\jbone\\Google Drive (1)\\Tweets for project\\'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]

json_files = sorted(json_files, key=lambda x: int("".join([i for i in x if i.isdigit()])))
```

```python
dataframe = pd.concat([pd.read_json(path_to_json+file) for file in json_files[1:10]])
```

```python
#dataframe = {}
frames=[]
for file in json_files[0:10]:
    print(file)
    current = pd.read_json(path_to_json+file)
    frames.append(current)
dataframe = pd.concat(frames)
```
